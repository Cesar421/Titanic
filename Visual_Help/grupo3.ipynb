```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

```python
x = np.linspace(-10, 10, 100)
y = np.cos(x)
y2 = np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(15, 10))
plt.plot(x, y, color="#7EF5A2", label="cos(x)", lw=10, ls=(0, (1, 1)))
plt.plot(x, y2, color="#7EE5F5", label="sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Grafico del coseno de x")
plt.legend()
```
