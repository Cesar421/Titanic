```python
from sklearn.datasets import make_blobs
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
plt.style.use("ggplot")
```

```python
X, y = make_blobs(n_samples=1000, n_features=2, centers=4)
```

```python
print(X.shape)
```

```python
print(y.shape)
```

```python
print(type(X))
```

```python
print(type(y))
```

```python
uniques, counts = np.unique(y, return_counts=True)
fig, ax = plt.subplots(1, 1, figsize=(7, 7))
ax.bar(uniques, counts)
ax.set_xticks(uniques);
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 7))
sns.kdeplot(X[:, 0], ax=ax[0])
ax[0].set_title("$x_1$")
sns.kdeplot(X[:, 1], ax=ax[1])
ax[1].set_title("$x_2$")
```

```python
fig, ax  = plt.subplots(1, 1, figsize=(10, 10))
ax.scatter(X[:, 0], X[:, 1], c=y, alpha=0.4)
ax.set_xlabel("$x_1$")
ax.set_ylabel("$x_2$")
```

```python
clf = LogisticRegression()
```

```python
clf
```

```python
clf.fit(X, y)
```

```python
point = np.array([[0, 0.], [1, 2], [-1, 1]])
point
```

```python
point.shape
```

```python
clf.predict(point)
```

```python
point = np.array([[-5, -8.]])
y_pred = clf.predict(point)
print(y_pred)
```

```python
fig, ax  = plt.subplots(1, 1, figsize=(10, 10))
ax.scatter(X[:, 0], X[:, 1], c=y_colors, alpha=0.4)
ax.scatter(point[:, 0], point[:, 1], marker="*", c=colors[y_pred], s=200)
ax.set_xlabel("$x_1$")
ax.set_ylabel("$x_2$")
```
